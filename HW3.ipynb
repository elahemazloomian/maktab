```python
import seaborn as se
import matplotlib.pyplot as plt
selected_columns = ["Outcome", "Insulin", "Age", "BMI", "BloodPressure", "Glucose"]
df_selected = df[selected_columns]
correlation_matrix = df_selected.corr()
print(correlation_matrix)
print("========================================")

plt.figure(figsize=(10, 8))
se.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0,fmt='.2f')


plt.title("correlation matrix")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()




outcome_corr = correlation_matrix['Outcome'].drop('Outcome')
max_corr_feature = outcome_corr.idxmax()
max_corr_value = outcome_corr.max()

print(f"• ویژگی: {max_corr_feature}")
print(f"• مقدار همبستگی: {max_corr_value:.4f}")

print("----------------------------------")
print("\nهمبستگی تمام ویژگی‌ها با Outcome:")
print(outcome_corr)
```
