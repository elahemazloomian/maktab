```python
import pandas as pd
import math

df = pd.read_csv("iris_dataset.csv")

df = df[["sepal length (cm)", "sepal width (cm)", "target"]]

p_test = df.loc[149]
p_train = df.drop(149)

pt = (p_test["sepal length (cm)"], p_test["sepal width (cm)"])

distances = []

for index, row in p_train.iterrows():
    p = (row["sepal length (cm)"], row["sepal width (cm)"])
    d = math.dist(pt, p)
    distances.append((index, d))

distances.sort(key=lambda x: x[1])
k1_index = distances[0][0]
k1_predicted_class = p_train.loc[k1_index, "target"]

print("---- K = 1 ----")
print("Closest neighbor index:", k1_index)
print("Predicted class:", k1_predicted_class)
print("Real class:", p_test["target"])
print("Correct?", k1_predicted_class == p_test["target"])

k3_indices = [distances[0][0], distances[1][0], distances[2][0]]

k3_classes = [
    p_train.loc[k3_indices[0], "target"],
    p_train.loc[k3_indices[1], "target"],
    p_train.loc[k3_indices[2], "target"]
]

print("\n---- K = 3 ----")
print("3 nearest indices:", k3_indices)
print("Their classes:", k3_classes)

pred_k3 = max(set(k3_classes), key=k3_classes.count)

print("Predicted class (k=3):", pred_k3)
print("Real class:", p_test["target"])
print("Correct?", pred_k3 == p_test["target"])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("iris_dataset.csv")
features = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
p_test = df.loc[149, features]
p_train = df.drop(149)[features]
target_train = df.drop(149)["target"]

distances = np.sqrt(((p_train - p_test) ** 2).sum(axis=1))
k3_indices = distances.nsmallest(3).index
k3_classes = df.loc[k3_indices, "target"]
pred_k3_no_norm = k3_classes.mode()[0]

print("---- K = 3    ----")
print("3 nearest indices:", list(k3_indices))
print("Their classes:", list(k3_classes))
print("Predicted class:", pred_k3_no_norm)
print("Real class:", df.loc[149, "target"])
print("Correct?", pred_k3_no_norm == df.loc[149, "target"])

mean_train = p_train.mean()
std_train = p_train.std()
train_norm = (p_train - mean_train) / std_train
test_norm = (p_test - mean_train) / std_train

distances_norm = np.sqrt(((train_norm - test_norm) ** 2).sum(axis=1))
k3_indices_norm = distances_norm.nsmallest(3).index
k3_classes_norm = df.loc[k3_indices_norm, "target"]
pred_k3_norm = k3_classes_norm.mode()[0]

print("\n---- K = 3    ----")
print("3 nearest indices:", list(k3_indices_norm))
print("Their classes:", list(k3_classes_norm))
print("Predicted class:", pred_k3_norm)
print("Real class:", df.loc[149, "target"])
print("Correct?", pred_k3_norm == df.loc[149, "target"])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# 1.  
# ---------------------------
train_data = pd.read_csv('iris_dataset.csv')
test_data = pd.read_csv('iris_test_samples.csv')

X_train = train_data.iloc[:, :-1].values
Y_train = train_data.iloc[:, -1].values

X_test = test_data.iloc[:, :-1].values
Y_test = test_data.iloc[:, -1].values

classes = np.unique(Y_train)

# ---------------------------
# 2.      
# ---------------------------
feature_stats = {}
for c in classes:
    data_c = X_train[Y_train == c]
    stats = {}
    for i in range(X_train.shape[1]):
        feature = data_c[:, i]
        stats[i] = {
            'mean': np.mean(feature),
            'var': np.var(feature),
            'min': np.min(feature),
            'max': np.max(feature),
            'range': np.max(feature) - np.min(feature)
        }
    feature_stats[c] = stats

print("Feature stats per class:")
for c in feature_stats:
    print(f"Class {c}: {feature_stats[c]}")

# ---------------------------
# 3.   
# ---------------------------
def gaussian_prob(x, mean, var):
    return (1.0 / np.sqrt(2 * np.pi * var)) * np.exp(-((x - mean) ** 2) / (2 * var))

# ---------------------------
# 4.    
# ---------------------------
def classify_single_feature(feature_index, X, verbose=False):
    predictions = []
    for idx, x in enumerate(X[:, feature_index]):
        probs = []
        for c in classes:
            mean = feature_stats[c][feature_index]['mean']
            var = feature_stats[c][feature_index]['var']
            probs.append(gaussian_prob(x, mean, var))
        pred_class = np.argmax(probs)
        predictions.append(pred_class)
        if verbose:
            print(f"Sample {idx}: x={x}, Probabilities={probs}, Predicted Class={pred_class}")
    return np.array(predictions)

y_pred_train_f3 = classify_single_feature(2, X_train)
precision_train_f3 = np.sum(y_pred_train_f3 == Y_train) / len(Y_train) * 100
print(f"\nTrain Precision using feature 3: {precision_train_f3:.2f}%")

y_pred_test_f3 = classify_single_feature(2, X_test, verbose=True)
precision_test_f3 = np.sum(y_pred_test_f3 == Y_test) / len(Y_test) * 100
print(f"\nTest Precision using feature 3: {precision_test_f3:.2f}%")

# ---------------------------
# 5.    
# ---------------------------
def classify_two_features(feature_indices, X, verbose=False):
    predictions = []
    for idx, x in enumerate(X[:, feature_indices]):
        probs_per_class = []
        for c in classes:
            probs = []
            for i, val in enumerate(x):
                mean = feature_stats[c][feature_indices[i]]['mean']
                var = feature_stats[c][feature_indices[i]]['var']
                probs.append(gaussian_prob(val, mean, var))
            probs_per_class.append(np.mean(probs))
        pred_class = np.argmax(probs_per_class)
        predictions.append(pred_class)
        if verbose:
            print(f"Sample {idx}: x={x}, Class Probabilities={probs_per_class}, Predicted Class={pred_class}")
    return np.array(predictions)

y_pred_test_f34 = classify_two_features([2,3], X_test, verbose=True)
precision_test_f34 = np.sum(y_pred_test_f34 == Y_test) / len(Y_test) * 100
print(f"\nTest Precision using features 3 & 4: {precision_test_f34:.2f}%")

# ---------------------------
# 6.    4      
# ---------------------------
def classify_all_features(X, verbose=False):
    predictions = []
    for idx, x in enumerate(X):
        probs_per_class = []
        for c in classes:
            probs = []
            for i, val in enumerate(x):
                mean = feature_stats[c][i]['mean']
                var = feature_stats[c][i]['var']
                probs.append(gaussian_prob(val, mean, var))
            top_two = sorted(probs, reverse=True)[:2]
            probs_per_class.append(np.mean(top_two))
        pred_class = np.argmax(probs_per_class)
        predictions.append(pred_class)
        if verbose:
            print(f"Sample {idx}: x={x}, Class Probabilities (top2 avg)={probs_per_class}, Predicted Class={pred_class}")
    return np.array(predictions)

y_pred_test_all = classify_all_features(X_test, verbose=True)
precision_test_all = np.sum(y_pred_test_all == Y_test) / len(Y_test) * 100
print(f"\nTest Precision using all features (top 2 probs): {precision_test_all:.2f}%")

# ---------------------------
# 7.    
# ---------------------------
color_map = {0: 'blue', 1: 'black', 2: 'red'}
plt.figure(figsize=(8,6))
for c in classes:
    plt.scatter(X_train[Y_train==c, 2], X_train[Y_train==c, 3],
                color=color_map[c], label=f"Class {c}")
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Classes (Train Data)')
plt.legend()
plt.show()
```
