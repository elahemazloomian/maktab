```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
import seaborn as sns

california = fetch_california_housing()
df=pd.DataFrame(data=california.data,columns=california.feature_names)
df['MedHouseVal']=california.target
print("---------------------california dataset---------------------")
display(df)

print("-------------------dimentions--------------------------")
#print(df.shape)
display(df.shape)
print("-----------------range of data in each column--------------------")
column_range=df.max()-df.min()
display(column_range)
print("----------------------test---------------------")
print(df["MedInc"].max())
print(df["MedInc"].min())
print("------------------------info-------------------")
display(df.info())
display(df.describe())
```

```python
cov_california=df.cov()
display(cov_california)
```

```python
corr_california=df.corr()
display(corr_california)
sns.heatmap(corr_california, annot=True)
```

```python
comparison = pd.DataFrame({
    "Covariance_with_price": cov_california["MedHouseVal"],
    "Correlation_with_price": corr_california["MedHouseVal"]
})

display(comparison)
```

```python
corr_california_matrix=np.corrcoef(corr_california)

pairs=[]

for i in range(len(corr_california_matrix)):
    for j in range(i+1,len(corr_california)):
        pairs.append((corr_california.iloc[i,j],df.columns[i],df.columns[j]))

sorted_pairs=sorted(pairs,key=lambda x: x[0])

print("------------------sorted pairs----------------------")
display(sorted_pairs)
print("SMALLEST VALUES : ")
print(sorted_pairs[:2])
print("LARGEST VALUES : ")
print(sorted_pairs[-2:])
```

```python
from sklearn.datasets import load_breast_cancer
from numpy.linalg import inv


breast_cancer=load_breast_cancer()
df=pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
df["target"]=breast_cancer.target
display(df)
display(df.shape)
display(df.info())
```

```python
print("MALIGNANT CLASS :")
malignant_class=df[df["target"]==0]
malignant_distance=malignant_class.head(3)
display(malignant_distance)
print("BENIGN CLASS :")
benign_class=df[df["target"]==1]
benign_distances=benign_class.head(3)
display(benign_distances)
```

```python
def Euclidean_distance(x, y):
    return np.sqrt(np.sum((x - y) ** 2))

def Manhattan_distance(x, y):
    return np.sum(np.abs(x - y))

def Chebyshev_distance(x, y):
    return np.max(np.abs(x - y))

def Cosine_distance(x, y):
    return 1 - (np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y)))

def Mahalanobis_distance(x, y, data):
    cov_matrix = np.cov(data.T)
    inv_cov = inv(cov_matrix)
    diff = x - y
    return np.sqrt(np.dot(np.dot(diff.T, inv_cov), diff))

data_features = df.drop(columns=["target"]).values

rows = []
feature_cols = breast_cancer.feature_names

for i in range(len(benign_distances)):
    for j in range(len(malignant_distance)):
        x = benign_distances.iloc[i][feature_cols].values.astype(float)
        y = malignant_distance.iloc[j][feature_cols].values.astype(float)

        rows.append({
            "Benign_idx": i,
            "Malignant_idx": j,
            "Euclidean": Euclidean_distance(x, y),
            "Manhattan": Manhattan_distance(x, y),
            "Chebyshev": Chebyshev_distance(x, y),
            "Cosine": Cosine_distance(x, y),
            "Mahalanobis": Mahalanobis_distance(x, y, data_features)
        })


distance_table = pd.DataFrame(rows)
print("\nDistance Comparison Table:")
display(distance_table)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from IPython.display import display

data = load_diabetes()
X = data.data
y = data.target
feature_names = data.feature_names

df_raw = pd.DataFrame(X, columns=feature_names)
display(df_raw.head())

#STANDARDIZATION:
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

df_std = pd.DataFrame(X_std, columns=feature_names)

summary = pd.DataFrame({
    "mean": df_std.mean(),
    "std": df_std.std()
})

print("--------------------------MEAN AND STANDARD DEVIATION--------------------------")
display(summary)

```

```python
pca = PCA()
pca.fit(X_std)

#explain variance
explained = pca.explained_variance_ratio_
cumulative = explained.cumsum()

pca_table = pd.DataFrame({
    "Explained Variance": explained,
    "Cumulative Variance": cumulative
})
display(pca_table)

# cumulative var plot
plt.figure(figsize=(7,4))
plt.plot(range(1, 11), cumulative, marker="o")
plt.xlabel(" number of items")
plt.ylabel("cum variance")
plt.title("cum variance/pca plot")
plt.grid(True)
plt.show()

n80 = np.argmax(cumulative >= 0.80) + 1
n90 = np.argmax(cumulative >= 0.90) + 1

print(f"80% : {n80}")
print(f"90% : {n90}\n")

```

```python

print("-----------------------main items--------------------------------")

loadings = pd.DataFrame(
    pca.components_[:3],
    index=["PC1", "PC2", "PC3"],
    columns=feature_names
)

print("3 frist items loads: ")
display(loadings)


```

```python
X_pca = pca.transform(X_std)
pc1 = X_pca[:, 0]
pc2 = X_pca[:, 1]



sc = plt.scatter(pc1, pc2, c=y, cmap='viridis', s=30)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA plot")
plt.figure(figsize=(7, 5))
plt.colorbar(sc, label="progress (goal)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
from sklearn.datasets import fetch_lfw_people


lfw = fetch_lfw_people(min_faces_per_person=20, resize=0.5, download_if_missing=True)
images = lfw.images
target_names = lfw.target_names
dataset_name = "LFW"
print("Loaded LFW dataset.")


# Dataset info
n_samples, h, w = images.shape
print("Number of images:", n_samples)
print("Image size:", h, "x", w)

# Create data matrix X: (n_samples , n_pixels)
#changing images to 2d vectors
X = images.reshape(n_samples, -1)
print("Data matrix shape:", X.shape)
```

```python
# Center data
X_mean = X.mean(axis=0)
X_centered = X - X_mean

# PCA with many components
n_components = min(150, n_samples)
pca = PCA(n_components=n_components, svd_solver='randomized')
pca.fit(X_centered)

components = pca.components_
explained = pca.explained_variance_ratio_

# Show first 4 PCs as images
fig, axes = plt.subplots(1,4, figsize=(12,4))
for i, ax in zip(range(4), axes):
    ax.imshow(components[i].reshape(h,w), cmap='gray')
    ax.set_title(f"PC{i+1}")
    ax.axis('off')
plt.show()
```

```python
idx = 0
orig = X[idx]

ks = [20, 80, 150]
recons = {}
mses = {}

for k in ks:
    pca_k = PCA(n_components=k, svd_solver='randomized')
    pca_k.fit(X_centered)

    coeffs = pca_k.transform((orig - X_mean).reshape(1,-1))
    recon = pca_k.inverse_transform(coeffs) + X_mean

    recons[k] = recon.reshape(h,w)
    mses[k] = mean_squared_error(orig, recon.flatten())

# Plot
fig, axes = plt.subplots(1,4, figsize=(15,4))
axes[0].imshow(orig.reshape(h,w), cmap='gray')
axes[0].set_title("Original")
axes[0].axis('off')

for ax, k in zip(axes[1:], ks):
    ax.imshow(recons[k], cmap='gray')
    ax.set_title(f"k={k}\nMSE={mses[k]:.4f}")
    ax.axis('off')

plt.show()
```

```python
sigma = 0.2
orig_norm = orig / orig.max()
noisy = orig_norm + np.random.normal(scale=sigma, size=orig_norm.shape)
noisy = np.clip(noisy, 0, 1)

X_norm = X / X.max()
Xn_centered = X_norm - X_norm.mean(axis=0)

pca_dn = PCA(n_components=150, svd_solver='randomized')
pca_dn.fit(Xn_centered)

coeffs_noisy = pca_dn.transform((noisy - X_norm.mean(axis=0)).reshape(1,-1))
recon_noisy = pca_dn.inverse_transform(coeffs_noisy)
recon_noisy = recon_noisy.reshape(h,w)

# Plot
fig, axes = plt.subplots(1,3, figsize=(12,4))
axes[0].imshow(orig_norm.reshape(h,w), cmap='gray'); axes[0].set_title("Original"); axes[0].axis('off')
axes[1].imshow(noisy.reshape(h,w), cmap='gray'); axes[1].set_title("Noisy"); axes[1].axis('off')
axes[2].imshow(recon_noisy, cmap='gray'); axes[2].set_title("Denoised (PCA)"); axes[2].axis('off')
plt.show()
```
